##  1. Create a list named products containing at least 6 product names

```python
products = ["Laptop", "Mouse", "Keyboard", "Monitor", "Headphones", "Printer"]
print(type(products))
print(products)
```

```python
sample_product = ("Laptop", 75000, "Electronics")
print(type(sample_product))
print(sample_product)
```

## 3. Print the 2nd and last product from the products list
####  i use slicing method for this 

```python
print("Second product:", products[1])
print("Last product:", products[-1])
```

## 4. Append two new product names and print updated list
#### i user append method , append method add new value in list 

```python
products.append("Webcam")
products.append("Speaker")

print("Updated products list:", products)
```
