# task :- 3

## 1. Create a dictionary with at least 6 entries

#### create dictionary and insert six product 

```python
price_dict = {
    "Laptop": 75000,
    "Monitor": 12000,
    "Keyboard": 1500,
    "Mouse": 800,
    "Webcam": 3500,
    "Headset": 2500
}
print(price_dict)
```

## 2. Code blocks for updates Add a new product

#### for this  i use key name for it we add new value

```python
price_dict["USB Hub"] = 1200 
print(price_dict)
```

#### Update existing product price

```python
price_dict["Monitor"] = 11500
print(price_dict)
```

#### Remove a product (handling non-existent cases)

```python
product_to_remove = "Speakers"
price_dict.pop(product_to_remove, None)  
```

```python
print(price_dict)
```

## 3. Print average price

```python
total_price = sum(price_dict.values())
average_price = total_price / len(price_dict)
print(f"Average Price: {average_price}")
```

## Extra: Print max and min price products

```python
max_product = max(price_dict, key=price_dict.get)
min_product = min(price_dict, key=price_dict.get)
print(f"Max Price: {max_product} ({price_dict[max_product]})")
print(f"Min Price: {min_product} ({price_dict[min_product]})")
```

## Assuming we have these from previous tasks 

```python
products = ["Laptop", "Monitor", "Keyboard", "Mouse", "Webcam", "Headset"]
categories = ["Electronics", "Electronics", "Accessories", "Accessories", "Electronics", "Audio"]
```

## 1. Create a list of tuples (catalog)

```python
catalog = []
for i in range(len(products)):
    name = products[i]
    price = price_dict.get(name, 0)
    category = categories[i]
    catalog.append((name, price, category))
print(i) 
```

## 2. Map categories to lists of product names

```python
category_to_products = {}
for name, price, category in catalog:
    if category not in category_to_products:
        category_to_products[category] = []
    category_to_products[category].append(name)

print(f"Category Map: {category_to_products}")
```

## 3. Print products from the largest category

```python
max_cat = max(category_to_products, key=lambda k: len(category_to_products[k]))
print(f"Category with most products: {max_cat}")
print(f"Products in {max_cat}: {category_to_products[max_cat]}")
```
